# Importing Dataset of Movies Between 2010-2019

```python
movie_b = pd.read_csv('hindi/bollywood_2010-2019.csv')
movie_a = pd.read_csv('hindi/bollywood_1990-2009.csv')
```

```python
movies = pd.concat([movie_a, movie_b])
movies
```

```python
meta_a = pd.read_csv('bollywood_meta_1990-2009.csv')
meta_b = pd.read_csv('bollywood_meta_2010-2019.csv')
meta = pd.concat([meta_a, meta_b])
```

```python
movies = pd.concat([movies, meta])
```

```python
movies
```
